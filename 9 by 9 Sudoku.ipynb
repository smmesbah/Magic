```python
import numpy as np
import time

def print_board(board):
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("-" * 21)

        for j in range(9):
            if j % 3 == 0 and j != 0:
                print("|", end=" ")
            print(board[i][j], end=" ")
        print()

def is_valid(board, row, col, num):
    if num in board[row] or num in [board[i][col] for i in range(9)]:
        return False

    start_row, start_col = 3 * (row // 3), 3 * (col // 3)
    for i in range(3):
        for j in range(3):
            if board[start_row + i][start_col + j] == num:
                return False

    return True

def solve_sudoku(board):
    empty = find_empty_location(board)

    if not empty:
        return True

    row, col = empty

    numbers = np.arange(1, 10)
    np.random.shuffle(numbers)

    for num in numbers:
        if is_valid(board, row, col, num):
            board[row][col] = num

            if solve_sudoku(board):
                return True

            board[row][col] = 0

    return False

def find_empty_location(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return (i, j)
    return None

# Create an empty Sudoku board
sudoku_board = np.zeros((9, 9), dtype=int)

tic = time.time()
if solve_sudoku(sudoku_board):
    toc = time.time()
    print("Solved Sudoku:")
    print_board(sudoku_board)
    print()
    print(f"Time taken: {toc - tic:.6f} seconds")
else:
    print("No solution exists.")
```
